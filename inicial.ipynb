### Case - Cancelamento de Clientes

Você foi contratado por uma empresa com mais de 800 mil clientes para um projeto de Dados. Recentemente a empresa percebeu que da sua base total de clientes, a maioria são clientes inativos, ou seja, que já cancelaram o serviço.

Precisando melhorar seus resultados ela quer conseguir entender os principais motivos desses cancelamentos e quais as ações mais eficientes para reduzir esse número.

```python
# Passo a passo do projeto
# 1 Importar a base de dados de clientes
# 2 Visualizar a base de dados
# 3 Corrigir as cagadas da base de dados
# 4 Analise dos cancelamentos
# 5 Analise da causa do cancelamento
# trabalhar com bd -> pandas
# graficos e base de dados -> plotly

```

```python
import pandas as pd

tabela = pd.read_csv("cancelamentos_sample.csv")

tabela = tabela.drop(columns="CustomerID")
display(tabela)
```

```python
display(tabela.info())
tabela = tabela.dropna()
display(tabela.info())
```

```python
tabela["cancelou"].value_counts()
tabela["cancelou"].value_counts(normalize=True)
# tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format)
```

```python
import plotly.express as px

grafico = px.histogram(tabela, x="idade")

grafico.show()
```
